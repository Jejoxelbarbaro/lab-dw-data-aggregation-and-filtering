This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by first performing data cleaning, formatting, and structuring.

```python
import numpy as np
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
marketing_customer_analysis = pd.read_csv(url)
marketing_customer_analysis.head()
```

```python
mkt_analysis_df = marketing_customer_analysis.copy()
mkt_analysis_df.info()
```

```python
mkt_analysis_df.columns = [column_name.lower().replace(" ", "_") for column_name in mkt_analysis_df.columns]
```

```python
mkt_analysis_df.isna().sum()
```

```python
mkt_analysis_df = mkt_analysis_df.dropna(subset=["state", "months_since_last_claim" ])
```

1. Create a new DataFrame that only includes customers who:
   - have a **low total_claim_amount** (e.g., below $1,000),
   - have a response "Yes" to the last marketing campaign.

```python
# total_claim_amount < 1000 FILTRAR VALORES DE LA COLUMNA EN UN NUEVO DF
clients_under_1k_df = mkt_analysis_df[
    (mkt_analysis_df["total_claim_amount"] < 1000) &
    (mkt_analysis_df["response"].str.lower() == "yes") ]
clients_under_1k_df.sample(5)
```

2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
mkt_analysis_yes_df = mkt_analysis_df[mkt_analysis_df["response"].str.lower() == "yes" ]
mkt_analysis_yes_df.head()
```

```python
mkt_analysis_yes_df["monthly_premium_auto"].mean()
```

```python
monthly_premium_gender = mkt_analysis_yes_df.pivot_table(
    values="monthly_premium_auto",       # the main information we want to display!
    index="policy_type",    # what becomes the rows
    columns="gender",     # what becomes the columns
)

monthly_premium_gender
```

```python
monthly_premium_gender_ltv = mkt_analysis_yes_df.pivot_table(
    values="customer_lifetime_value",       # the main information we want to display!
    index="policy_type",    # what becomes the rows
    columns="gender",     # what becomes the columns
)

monthly_premium_gender_ltv
```

```python
total_claim_inv = mkt_analysis_yes_df.pivot_table(
    values=["monthly_premium_auto","customer_lifetime_value", "total_claim_amount" ], 
    index="policy_type", 
    columns="gender",   
)

total_claim_inv
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
customer_state = mkt_analysis_df.groupby("state").size()
customer_state = customer_state[customer_state > 1000]
customer_state
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
ltv_analyze = mkt_analysis_yes_df.pivot_table(
    values="customer_lifetime_value",
    index="education", 
    columns="gender",
    aggfunc = ["min", "max", "median" ]
)

ltv_analyze
```
